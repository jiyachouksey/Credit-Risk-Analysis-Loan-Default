INCOME V/S DEFAULT


```python
df.groupby('default')['annual_income'].mean()
```

```python
sns.boxplot(x='default', y='annual_income', data=df)
```

Insight: **Defaulters tend to have lower average income compared to non-defaulters.**

Loan Amount v/s Default

```python
df.groupby('default')['loan_amount'].mean()
```

Credit History v/s default

```python
df.groupby('default')['credit_history_years'].mean()
```

Past Default Impact

```python
pd.crosstab(df['past_Default'], df['default'], normalize='index') * 100
```

Loan Purpose Risk

```python
pd.crosstab(df['loan_purpose'], df['default'], normalize='index') * 100
```

Compare key metrics

```python
df.groupby('default')[[
    'annual_income',
    'loan_amount',
    'interest_rate',
    'credit_history_years',
    'loan_to_income_ratio'
]].mean()
```

Correlation check

```python
df[['annual_income',
    'loan_amount',
    'interest_rate',
    'credit_history_years',
    'loan_to_income_ratio',
    'default']].corr()
```

**Loan-to-income ratio and credit history showed noticeable correlation with default risk**

Loan-to-income ratio

```python
df['risk_segment'] = pd.cut(
    df['loan_to_income_ratio'],
    bins=[0, 0.1, 0.2, 1],
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)
```

```python
pd.crosstab(df['risk_segment'], df['default'], normalize='index') * 100
```

High loan-to-income customers show significantly higher default rates

```python
pd.crosstab(df['credit_grade'], df['default'], normalize='index') * 100
```
